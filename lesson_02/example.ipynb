```python
from queue import Queue

q = Queue(maxsize=3)

q.put('a')
q.put('b')
print(q.full())
q.put('c')
print(q.full())

if not q.full():
  q.put('d')
else:
  el = q.get()
  q.put('d')

print(' '.join((q.queue)))
```

```python
from collections import deque

d = deque()
print(d)  # deque([])
d.append('right')
d.appendleft('left')

d.extend(['a', 'b', 'c'])
d.extendleft(['x', 'y', 'z'])
print(d)
d.rotate(-2)
print(d)
```

```python
result = [23, 11, 12, 32, 10, 25, 65, 43, 42, 89, 100, 3]

def solve(data: list) -> list:
  dq = deque()
  dq.append(0)
  for i in data:
    if i % 2 == 0:
      dq.append(i)
    else:
      dq.appendleft(i)
  return list(dq)


r = solve(result)
print(r)
```
